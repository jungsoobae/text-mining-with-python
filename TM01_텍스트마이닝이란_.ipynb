#예제1_텍스트 데이터 전처리 (정규표현식 활용)

```python
import re

text ='텍스트마이닝은 데이터를 분석하는 기술이다!!!'
clean_text = re.sub('[^가-힣 ]','',text)
print(clean_text)
```

```python
from konlpy.tag import Okt
okt=Okt()

text ="텍스트마이닝은 데이터를 분석하는 기술이다"
morphs = okt.morphs(text)
print(morphs)
```
