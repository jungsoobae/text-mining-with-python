#예제5)	파이썬의 산술 연산자를 연습해 본다.

```python
a = 15
b = 3

print('a + b =', a + b)
print('a - b =', a - b)
print('a * b =', a * b)
print('a / b =', a / b)
print('a // b =', a // b)
print('a % b =', a % b)
print('a ** b =', a ** b)
```
