#예제11) 3차원 배열 요소 접근

```python
import numpy as np
arr = np.array([10, 20, 30, 40, 50, 60])
# 기본 슬라이싱
print(arr[1:4]) # [20 30 40]
print(arr[:3]) # [10 20 30] (start 생략)
print(arr[3:]) # [40 50 60] (end 생략)
print(arr[::2]) # [10 30 50] (step 지정)
print(arr[::-1]) # [60 50 40 30 20 10] (역순)
```

#예제26) 텍스트 데이터 정리하기

#예제27) 문장의 단어 길이 분석하기

```python
import numpy as np
sentence = "자연어 처리는 인공지능에서 중요한 분야 중 하나이다"
words = np.array(sentence.split()) # 공백을 기준으로 단어 분리
# (1) 각 단어의 길이 계산
word_lengths = np.array([len(word) for word in words])
# (2) 통계 분석 수행
mean_length = np.mean(word_lengths) # 평균 단어 길이
median_length = np.median(word_lengths) # 중앙값
std_length = np.std(word_lengths) # 표준편차
freq_length = np.bincount(word_lengths) # 단어 길이별 빈도
# (3) 결과 출력
print("단어 목록:", words)
print("단어 길이:", word_lengths)
print("평균 단어 길이:", mean_length)
print("중앙값:", median_length)
print("표준편차:", std_length)
print("단어 길이별 빈도:", freq_length)
```

#예제28) 단어 길이 분석과 조건 기반 단어 추출

```python
import numpy as np
sentence = "텍스트 마이닝을 활용하여 데이터에서 의미를 추출한다"
words = np.array(sentence.split()) # 공백 기준으로 단어 분리
# (1) 각 단어의 길이 계산
word_lengths = np.array([len(word) for word in words])
# (2) 가장 긴 단어 찾기
longest_index = np.argmax(word_lengths) # 최댓값 위치
longest_word = words[longest_index]
# (3) 가장 짧은 단어 찾기
shortest_index = np.argmin(word_lengths) # 최솟값 위치
shortest_word = words[shortest_index]
# (4) 길이가 4 이상인 단어의 위치 찾기 (np.where() 사용)
long_words_indices = np.where(word_lengths >= 4)
long_words = words[long_words_indices]
# (5) 특정 단어("텍스트")의 위치 찾기
text_word_indices = np.where(words == "텍스트")
# (6) 결과 출력
print("문장:", sentence)
print("단어 목록:", words)
print("단어 길이:", word_lengths)
print(f"가장 긴 단어: {longest_word} (길이: {word_lengths[longest_index]})")
print(f"가장 짧은 단어: {shortest_word} (길이: {word_lengths[shortest_index]})")
print("길이가 4 이상인 단어의 위치:", long_words_indices[0])
print("길이가 4 이상인 단어:", long_words)
print("특정 단어 '텍스트'의 위치:", text_word_indices[0])
```

#4.5.2. 미션2: 리뷰 데이터로 감성 분석 체험하기

```python
reviews = np.array([
 "이 제품은 정말 좋다. 품질이 만족스럽다.",
 "별로다. 가격 대비 성능이 별로 좋지 않다.",
 "매우 만족스럽고 품질도 좋다!",
 "불만족스럽다. 다시는 구매하지 않을 것이다.",
 "디자인이 예쁘고 사용감도 좋다."
])
```

```python
positive_words = np.array(["좋다", "만족", "예쁘"])
negative_words = np.array(["별로", "불만족"])
```

```python
positive_count = np.array([np.char.count(reviews, word) for word in positive_words])
```

```python
negative_count = np.array([np.char.count(reviews, word) for word in
negative_words])
```

```python
for i, review in enumerate(reviews):
    pos_words_in_review = [positive_words[j] for j in range(len(positive_words)) if positive_count[j][i] > 0]
    pos_word_list = ", ".join(pos_words_in_review) if len(pos_words_in_review) > 0 else "없음"
    neg_words_in_review = [negative_words[j] for j in range(len(negative_words)) if negative_count[j][i] > 0]
    neg_word_list = ", ".join(neg_words_in_review) if len(neg_words_in_review) > 0 else "없음"
    print(f"리뷰 {i+1}: {review}")
    print(f"긍정 단어 수: {np.sum(positive_count[:, i])} ({pos_word_list})")
    print(f"부정 단어 수: {np.sum(negative_count[:, i])} ({neg_word_list})\n")
```

```python
# 리뷰 리스트를 하나씩 살펴보자!
for i, review in enumerate(reviews):
    # 긍정 단어 찾기
    pos_words_in_review = []
    for j in range(len(positive_words)):
        if positive_count[j][i] > 0:
            pos_words_in_review.append(positive_words[j])

    # 부정 단어 찾기
    neg_words_in_review = []
    for j in range(len(negative_words)):
        if negative_count[j][i] > 0:
            neg_words_in_review.append(negative_words[j])

    # 긍정 단어 리스트를 문자열로 변환
    if len(pos_words_in_review) > 0:
        pos_word_list = ", ".join(pos_words_in_review)
    else:
        pos_word_list = "없음"

    # 부정 단어 리스트를 문자열로 변환
    if len(neg_words_in_review) > 0:
        neg_word_list = ", ".join(neg_words_in_review)
    else:
        neg_word_list = "없음"

    # 결과 출력
    print(f"리뷰 {i+1}: {review}")
    print(f"긍정 단어 수: {np.sum(positive_count[:, i])} ({pos_word_list})")
    print(f"부정 단어 수: {np.sum(negative_count[:, i])} ({neg_word_list})\n")
```
