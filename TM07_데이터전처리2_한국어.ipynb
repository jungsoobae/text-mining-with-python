```python
import re
from konlpy.tag import Okt
from collections import Counter
from nltk import ngrams
text = "*í•œêµ­ì–´ ìì—°ì–´ ì²˜ë¦¬ëŠ” ë³µì¡í•˜ì§€ë§Œ í¥ë¯¸ë¡œìš´ ë¶„ì•¼ì…ë‹ˆë‹¤!." # 1. ë°ì´í„° ìˆ˜ì§‘
# 2. í…ìŠ¤íŠ¸ ì •ì œ (íŠ¹ìˆ˜ë¬¸ì ì œê±°)
cleaned_text = re.sub(r"[^\w\s]", "", text)
# 3. í† í°í™”
okt = Okt()
tokens = okt.morphs(cleaned_text)
# 4. ì–´ê°„ì¶”ì¶œ
stemmed_tokens = okt.morphs(cleaned_text, stem=True)
# 5. í’ˆì‚¬ íƒœê¹…
pos_tags = okt.pos(cleaned_text)
# 6. ë¶ˆìš©ì–´ ì œê±°
stop_words = ['ëŠ”', 'ì˜', 'ì´', 'ê°€', 'ì—', 'ë¥¼', 'ì…ë‹ˆë‹¤']
filtered_tokens = [word for word in tokens if word not in stop_words]
# 7. ë‹¨ì–´ ë¹ˆë„ ê³„ì‚°
word_counts = Counter(filtered_tokens)
# 8. N-ê·¸ë¨ ìƒì„± í•¨ìˆ˜
bigrams = list(ngrams(filtered_tokens, 2))
print("í…ìŠ¤íŠ¸ ì •ì œ:", cleaned_text)
print("í† í°í™”:", tokens)
print("ì–´ê°„ ì¶”ì¶œ:", stemmed_tokens)
print("í’ˆì‚¬ íƒœê¹…:", pos_tags)
print("ë¶ˆìš©ì–´ ì œê±°:", filtered_tokens)
print("ë‹¨ì–´ ë¹ˆë„ ê³„ì‚°:", word_counts)
print("N-ê·¸ë¨ ìƒì„±:", bigrams)
```
