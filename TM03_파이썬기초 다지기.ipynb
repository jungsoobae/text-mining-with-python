#ì˜ˆì œ53)	ë‹¨ì–´ ê¸¸ì´ì— ë”°ë¥¸ ë¶„ë¥˜

```python
def classify_words(text):
    words = text.split()  # ì…ë ¥ëœ ë¬¸ì¥ì„ ê³µë°±ì„ ê¸°ì¤€ìœ¼ë¡œ ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
    short_words = []  # ì§§ì€ ë‹¨ì–´(4ì ë¯¸ë§Œ) ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
    long_words = []  # ê¸´ ë‹¨ì–´(4ì ì´ìƒ) ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
    for word in words:  # ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ë¥¼ ìˆœíšŒí•˜ë©´ì„œ ê¸¸ì´ë¥¼ ë¹„êµ
        if len(word) >= 4:  # ë‹¨ì–´ ê¸¸ì´ê°€ 4ì ì´ìƒì´ë©´ ê¸´ ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
            long_words.append(word)
        else:  # ë‹¨ì–´ ê¸¸ì´ê°€ 4ì ë¯¸ë§Œì´ë©´ ì§§ì€ ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
            short_words.append(word)
    # ê²°ê³¼ ì¶œë ¥
    print("- ì§§ì€ ë‹¨ì–´:", ", ".join(short_words))  # ì§§ì€ ë‹¨ì–´ ì¶œë ¥ (ì‰¼í‘œë¡œ êµ¬ë¶„)
    print("- ê¸´ ë‹¨ì–´:", ", ".join(long_words))  # ê¸´ ë‹¨ì–´ ì¶œë ¥ (ì‰¼í‘œë¡œ êµ¬ë¶„)
```

```python
sentence = """
ë³´ê³  ì‹¶ë‹¤ ì´ë ‡ê²Œ ë§í•˜ë‹ˆê¹Œ ë” ë³´ê³  ì‹¶ë‹¤
ë„ˆí¬ ì‚¬ì§„ì„ ë³´ê³  ìˆì–´ë„ ë³´ê³  ì‹¶ë‹¤
ë„ˆë¬´ ì•¼ì†í•œ ì‹œê°„ ë‚˜ëŠ” ìš°ë¦¬ê°€ ë°‰ë‹¤
ì´ì œ ì–¼êµ´ í•œ ë²ˆ ë³´ëŠ” ê²ƒë„ í˜ë“¤ì–´ì§„ ìš°ë¦¬ê°€
ì—¬ê¸´ ì˜¨í†µ ê²¨ìš¸ ë¿ì´ì•¼
8ì›”ì—ë„ ê²¨ìš¸ì´ ì™€
ë§ˆìŒì€ ì‹œê°„ì„ ë‹¬ë ¤ê°€ë„¤
í™€ë¡œ ë‚¨ì€ ì„¤êµ­ì—´ì°¨
ë„¤ ì† ì¡ê³  ì§€êµ¬ ë°˜ëŒ€í¸ê¹Œì§€ ê°€
ì—¬ê¸´ ë”°ëœ»í•œë° ë„ˆëŠ” ì € ë¶ìª½ ë•…ì— ì‚´ì•„
"""
classify_words(sentence)  # ì…ë ¥ëœ ê°€ì‚¬ë¡œ í•¨ìˆ˜ ì‹¤í–‰
```
