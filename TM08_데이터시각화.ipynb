##예제1)	라인 플롯(Line Plot)

```python
import matplotlib.pyplot as plt
# 데이터 정의
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
# 라인 플롯
plt.plot(x, y, label="Prime Numbers", color="blue", marker="o")
plt.title("Line Plot Example")   # 그래프 제목
plt.xlabel("X-axis")             # X축 레이블
plt.ylabel("Y-axis")             # Y축 레이블
plt.legend()                     # 범례 표시
plt.grid(True)                   # 그리드 추가
plt.show()                       # 그래프 출력
```

##예제2)	막대그래프-plt.bar()

```python
import matplotlib.pyplot as plt
# 데이터
categories = ["A", "B", "C", "D"]
values = [10, 15, 7, 20]
# 막대그래프 생성
plt.bar(categories, values, color="skyblue")
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

##예제3)	파이 차트 plt.pie()

```python
import matplotlib.pyplot as plt
# 데이터 정의
sizes = [30, 20, 50]
labels = ["A", "B", "C"]
colors = ["red", "blue", "green"]
explode = [0, 0.2, 0]  # 두 번째 조각 강조
# 파이 차트 그리기
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct="%1.1f%%", shadow=True, startangle=90)
# 제목 추가
plt.title("Pie Chart Example")
# 차트 표시
plt.show()
```

##예제4)	가우시안(정규)분포 히스토그램 그리기

##예제8)	타이타닉 데이터셋을 불러보자.

```python
import pandas as pd
import matplotlib.pyplot as plt
# 타이타닉 데이터셋 로드
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
print(titanic.head())
```

##예제9)	생존 여부 비율 원형 그래프 그리기

```python
# 생존 여부 비율
survival_counts = titanic['Survived'].value_counts()
labels = ['Died', 'Survived']
# 원형 그래프 그리기
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', colors=['red', 'blue'])
plt.title('Survival Rate on Titanic')
plt.show()
```

##예제10)	성별 생존률 계산

```python
# 성별 생존률 계산
gender_survival = titanic.groupby('Sex')['Survived'].mean()
# 성별 생존률 시각화
gender_survival.plot(kind='bar', color=['blue', 'pink'], alpha=0.7)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
```

##예제11)	나이에 따른 생존률 분석

```python
# 나이 결측값 처리 (중앙값으로 대체)
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())  # inplace=True 대신 직접 할당 방식 사용
# 어린이 여부 컬럼 생성
titanic['Child'] = titanic['Age'] <= 16  # 16세 이하는 True, 17세 이상은 False
# 어린이와 성인의 생존률 계산
age_survival = titanic.groupby('Child')['Survived'].mean()
# True/False를 'Adult', 'Child'로 변경
age_survival.index = ['Adult', 'Child']
# 생존률 시각화
age_survival.plot(kind='bar', color=['green', 'orange'], alpha=0.7, edgecolor='black')
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

##예제12)	좌석 등급별 생존률 분석

```python
# 좌석 등급별 생존률 계산
class_survival = titanic.groupby('Pclass')['Survived'].mean()
print("Survival Rate by Pclass:\n", class_survival)
# 막대 그래프 그리기
ax = class_survival.plot(kind='bar', color=['gold', 'silver', 'brown'], alpha=0.7)
plt.title('Survival Rate by Pclass')
plt.xlabel('Pclass (Seat Class)')
plt.ylabel('Survival Rate')
plt.grid(axis='y')
plt.xticks(rotation=0)
# 그래프 출력
plt.show()
```

##예제19)	NetworkX를 활용한 기본 네트워크 그래프 생성

```python
# 1. 필요한 라이브러리 불러오기
import networkx as nx  # 네트워크 그래프 생성
import matplotlib.pyplot as plt  # 그래프 시각화
```

```python
# 2. 네트워크 그래프 객체 생성
G = nx.Graph()
```

```python
# 3. 단어 (노드) 및 관계 (엣지) 추가
words = ["dog", "walk", "park", "lost", "find"]
edges = [("dog", "walk"), ("walk", "park"), ("park", "lost"), ("lost", "find"), ("dog", "find")]
G.add_nodes_from(words)  # 노드 추가
G.add_edges_from(edges)  # 엣지 추가
```

```python
# 4. 네트워크 그래프 분석
print("노드 개수:", G.number_of_nodes()) # 노드 개수 출력
print("엣지 개수:", G.number_of_edges()) # 엣지 개수 출력
print("노드 연결 상태:", G.degree()) # 각 노드별 연결 개수 출력
```

```python
# 5. 네트워크 그래프 시각화
plt.figure(figsize=(6,6))
nx.draw(G, with_labels=True, node_color="lightblue", edge_color="gray", font_size=12)
plt.show()
```
